```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
dangerDrop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
railGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
# parameters: start,end,increment
# axis has to be multiple of other one, 13 numbers, 130
x_axis = np.arange(0,130,10)

plt.title("Coaster Speed Over Time")

plt.plot(x_axis,dangerDrop,marker="o",label="Danger Drop")
plt.plot(x_axis,railGun,marker="x",label="Rail Gun")
plt.legend(loc="upper right")
plt.xlabel("Coaster Run Time")
plt.ylabel("Speed (MPH)")

plt.grid()
```

```python
plt.savefig('coaster.png')
```

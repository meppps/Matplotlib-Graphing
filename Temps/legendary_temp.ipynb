```python
# Include this line to make plots interactive
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
temps = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
x_axis = np.arange(1, 13 ,1)

x_axis

celTemps = [(x-32) * 0.56 for x in temps]
```

```python
plt.plot(x_axis, celTemps, marker="o", label="Celcius")
plt.plot(x_axis, temps, marker=">", label="Far")
plt.legend(loc="upper right")

plt.xlabel("Months")
plt.ylabel("Degrees")
```
